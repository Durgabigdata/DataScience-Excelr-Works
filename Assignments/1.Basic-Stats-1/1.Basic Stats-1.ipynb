## STANDARDIZATION

### Concept of Standardization

->The process of transforming data to have a mean of 0(zero) and a standard deviation of 1.

->Converting large or small values to standard values makes easy to compare with any features.

Z=(x-mu)/sigma

z-> z -score or standardized value

x->our x value

mu->mean

sigma->standard deviation

```python
df["Avg Price"].describe()
```

```python
df["Standardized_Avg_Price"]=(df["Avg Price"]-df["Avg Price"].mean())/df["Avg Price"].std()
```

```python
df["Standardized_Avg_Price"].describe()
```

### Comparison of Data Distributions

```python
fig=plt.figure(figsize=(15,8))
plt.subplot(1,2,1)
plt.hist(df["Avg Price"])
plt.title("BEFORE")

plt.subplot(1,2,2)
plt.hist(df["Standardized_Avg_Price"])
plt.title("AFTER")
plt.show()
```
