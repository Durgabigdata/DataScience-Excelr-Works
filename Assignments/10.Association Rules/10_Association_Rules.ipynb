### Importing Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori, association_rules

```

```python
data=pd.read_excel("/content/Online retail.xlsx",header=None)
```

## Understanding the Data and Preprocessing the data

Checking for Null Values

```python
data.isnull().sum()
```

### Checking for duplicates

```python
data[data.duplicated(keep=False)]
```

```python
data.drop_duplicates(keep="first",inplace=True) #dropping the duplicates
```

```python
data.reset_index(drop=True,inplace=True)
```

```python
data
```

```python
data['Items'] = data.iloc[:, 0].str.split(',')   # Split items based on the commas and keeping those in items column
```

```python
data.Items
```

```python
all_items = sorted(set(item for sublist in data['Items'] for item in sublist))  # Create a list of all unique items
```

```python
all_items
```

```python
all_items.remove(all_items[0])  #the item is present twice because of space in that item name.so dropping one of that item.
```

```python
all_items      #These all are unique items in the retail store
```

```python
df = pd.DataFrame(0, index=data.index, columns=all_items) # Create a new DataFrame where each item is a column and each transaction is a row
```

```python
df   #This is the new data frame
```

```python
for idx, items in enumerate(data['Items']):
    df.loc[idx, list(set(items))] = 1
```

### Checking for Null Values

```python
df.isnull().sum()
```

```python
df.drop(df.columns[-1], axis=1, inplace=True)
```

### Applying Apriori Algorithm

```python
frequent_itemsets = apriori(df, min_support=0.01,use_colnames=True)   #This apriori algorithm is useful to find the frequency of different items
frequent_itemsets
```

### Applying Association Rule

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1) #using metric Lift and keeping threshold at 1.
rules
```

By increasing the Lift value we can get the important items which are togetherly buying.

```python
len(rules[rules["lift"]>2])
```

## Finding the Top Item which bought together

```python
rules[rules["lift"]>3]
```

The whole wheat pasta and olive oil are the products which are buying very frequently by the most of customers.

```python
rules[(rules["confidence"]>0.5)&(rules["lift"]>2)]
```

we can say that Spaghetti will definitely bought while buying the frozen vegetables and ground beef.we can say with 51% confidence and lift value was 2.2 which was very positive association.
